#1. Write a query that calculates the difference between the highest salaries found in the human resource and operations departments. Output just the difference in salaries.

```python
#merging the two DataFrames using the inner join and merge()
db_data = pd.merge(left=db_employee, right=db_department, left_on='department_id', right_on='id')
db_data.head()
```

```python
#overview of the max salary in the departments
db_data.groupby('department').salary.max()
```

```python
#selecting the specific departments using the loc index
db_human_resource = db_data.loc[db_data.department == 'human resource']
db_operations = db_data.loc[db_data.department == 'operation']
```

```python
#find the max salary in each department assigning it to a variable
max_salary_HR = db_human_resource.salary.max()

max_salary_OP = db_operations.salary.max()

#find the difference between the max salary in the departments.
salary_diff = (max_salary_OP - max_salary_HR)
salary_diff
```
